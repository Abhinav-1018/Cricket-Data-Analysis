# BATTING SUMMARY

```python
data_bat = pd.read_csv("batting_summary.csv")
df_bat= pd.DataFrame(data_bat)
data_bat.head()
```

```python
merge_df= pd.merge(df_match, data_bat, on ='Match_no')
merge_df.head()
```

```python
merge_df["Out/Not Out"]= merge_df.Dismissal.apply(lambda x: "Not Out" if x=="not out" else "Out")
merge_df.head(11)
```

```python
df_mergebat= merge_df.drop(columns= ['Dismissal'])
df_mergebat.head(11)
```
